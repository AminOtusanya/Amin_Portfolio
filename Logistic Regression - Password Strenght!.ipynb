```python
x = [labels[0] for labels in password_tuple]
y = [labels[1] for labels in password_tuple]
```

```python
'Next we convert our data into characters'
```

```python
def world_divide_char(inputs):
    character=[]
    for i in inputs:
        character.append(i)
    return character     
```

```python
'Next we pass our data into the data TFidfvectorizer'
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer=TfidfVectorizer(tokenizer=world_divide_char)
```

```python
x=vectorizer.fit_transform(x)
```

```python
x.shape
```

```python
x
```

```python
vectorizer.get_feature_names_out()
```

```python
first_document_vector=x[0]
first_document_vector
```

```python
first_document_vector.T.todense()
```

```python
df=pd.DataFrame(first_document_vector.T.todense(),index = vectorizer.get_feature_names_out(),columns =['TF-IDF'])
df.sort_values(by=['TF-IDF'],ascending = False)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape 
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf=LogisticRegression(random_state=0,multi_class='auto')
```

```python
clf.fit(x_train,y_train)
```

```python
dt = np.array(['%@123abcd'])
pred=vectorizer.transform(dt)
clf.predict(pred)
```

```python
y_pred=clf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score 
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
